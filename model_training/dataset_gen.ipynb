```python
import numpy as np
import pandas as pd
import os
```

```python
fruits = pd.DataFrame()

fruits["weight"] = pd.NaT #weight in grams
fruits["peel color"] = pd.NaT #Hue value(0-360)
fruits["flesh color"] = pd.NaT #Hue value(0-360)
fruits["texture"] = pd.NaT #smooth = 0 rough = 100 
fruits["size"] = pd.NaT #size of the fruit in mm
fruits["name"] = pd.NaT #name of the fruit (mango, apple, pineapple, orange, unknown)
```

```python
fruits
```

```python
def interpolate_fruits(frame, weight, peel_color, flesh_color, texture, size, name, n_vals = 500):
    temp_fruits = frame
    for _ in range(n_vals):
        instance_weight = np.random.randint(weight.get("min"), weight.get("max"))
        instance_peel_color = np.random.randint(peel_color.get("min"), peel_color.get("max"))
        instance_flesh_color = np.random.randint(flesh_color.get("min"), flesh_color.get("max"))
        instance_texture = np.random.randint(texture.get("min"), texture.get("max"))
        instance_size = np.random.randint(size.get("min"), size.get("max"))

        fruit = pd.DataFrame({"weight": [instance_weight], "peel color":[instance_peel_color], "texture": [instance_texture], "flesh color": [instance_flesh_color], "size":[instance_size], "name":[name]})

        temp_fruits = pd.concat([temp_fruits, fruit], ignore_index=True, axis=0)
    return temp_fruits
```
